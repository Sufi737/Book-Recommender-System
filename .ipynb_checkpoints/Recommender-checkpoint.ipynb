```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
books = pd.read_csv('data/books_updated.csv')
books.head()
```

### We will require average rating of a book and the number of ratings given to a book. Let us see if there is a relationship between average rating of a book and the number of ratings given.

```python
plt.figure(figsize=(15,8))
plt.scatter(x=books['average_rating'], y=books['ratings_count'])
```

### There is a potential positive relation between average rating and ratings count as we can see if there is a high ratings count then average rating is also high

### We require a matrix which has association of all users with all books. If a user has not rated a book, the value will be null. We can use pivot table to create this matrix. This pivot table will contain index as user_id, columns as book title and values as rating given by the user.

```python
ratings = pd.read_csv('data/ratings.csv')
ratings.head()
```

```python
books_title = books[['book_id', 'title']]
books_title.head()
```

```python
book_ratings = pd.merge(ratings, books_title, on="book_id", how='left')
book_ratings.head()
```

```python
book_ratings.describe()
```

```python
book_matrix = book_ratings.pivot_table(index='user_id', columns='title', values='rating')
book_matrix.head()
```

### The pivot table is ready. Selecting 2 books at random as the books already read by the user. We will take 'A Brief History of Time' and '1984'. We have to first create seperate two dataframes for each movie which contains user ratings for these two movies

```python
nineteen84_rating = book_matrix['1984']
history_rating = book_matrix['A Brief History of Time']
```

```python
history_rating.head()
```

### We will now see which movies are highly correlated with the above two movies. We can use pandas corrwith functionality to achieve this.

```python
similar_to_nineteen84=book_matrix.corrwith(nineteen84_rating)
similar_to_history=book_matrix.corrwith(history_rating)
```

```python
similar_to_history
```

```python
corr_1984 = pd.DataFrame(similar_to_nineteen84, columns=['Correlation'])
corr_1984.dropna(inplace=True)
corr_1984.head()
```

```python
corr_history = pd.DataFrame(similar_to_history, columns=['Correlation'])
corr_history.dropna(inplace=True)
corr_history.head()
```

### The list of similar books is not satisfactory. This is because we do not have all the ratings data for each book with us. We have at max only 100 ratings given by users for a book. Also, we do not have ratings data for all the books, which motivates me to discard this method here.

### We can also try to find similar books using the tags given by users to books. 

```python
books['tag_name'].head()
```

```python
cv = CountVectorizer(tokenizer=lambda x: [i.strip() for i in x.split(',')], lowercase=False)
```

```python
count_matrix = cv.fit_transform(books['tag_name']) # actual count, output will be a sparse matrix
```

```python
count_matrix
```

```python
# generating the cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
cosine_sim
```

### The above matrix stores the cosine similarity of each book with all the other books. Notice the diagonals are 1 which means the book is most similar to itself. The structure of the matrix is as follows:

###          $~~~~~~~~~~~$      Book1  Book2  Book3 ...... BookN

###  Book1
###  Book2
###  ....
###  BookN

```python
indices = pd.Series(books['title'])
indices
```

```python
idx = indices[indices == '1984'].index[0]
```

```python
idx
```

```python
matrix_row = pd.Series(cosine_sim[idx])
```

```python
matrix_row
```

### The above series contains the cosine similarity of book '1984' with the rest of the books. We can sort this list according to the value of cosine similarity to get the indexes of the most similar books. 

```python
matrix_row = matrix_row.sort_values(ascending=False)
```
